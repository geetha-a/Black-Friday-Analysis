```python
import pandas as pd
```

```python
df = pd.read_csv("BlackFriday (1).csv")
```

```python
df.groupby('Gender').size()
```

```python
df.groupby("Gender").size().plot(kind="pie",autopct= ("%.1f") ,figsize=(5,5))
```

```python
df.groupby("Gender").sum()["Purchase"].plot(kind = "pie" ,autopct ="%0.1f",)
```

```python
df.groupby('Gender').sum()['Purchase'].plot(kind = 'pie', autopct = "%0.1f")
```
